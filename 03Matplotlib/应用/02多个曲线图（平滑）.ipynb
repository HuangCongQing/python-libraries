```python
# -*- coding:utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
# x = np.linspace(0, 2 * (np.pi))  #numpy.linspace(开始，终值(含终值))，个数)
# y1 = np.sin(x)
# y2 = np.cos(x)
x = [500,750,1000,1250,1500,1750,2000,2500]
y0 = [0.45, 0.63, 0.45, 0.63, 0.45, 0.63, 0.45, 0.63]
y1 = [0.45, 0.63, 0.78, 0.63, 0.89, 0.922, 0.98, 0.99]


# https://www.cnblogs.com/xyou/p/8578896.html
#插值法之后的x轴值，表示从0到10间距为0.5的200个数
xnew = np.arange(500, 2500, 100)

#实现函数
func1 = interpolate.interp1d(x, y0, kind='cubic')
func2 = interpolate.interp1d(x, y1, kind='cubic')

#利用xnew和func函数生成ynew,xnew数量等于ynew数量
ynew1 = func1(xnew)
ynew2 = func2(xnew)

#常见线的属性有：color,label,linewidth,linestyle,marker等
plt.grid()  # 生成网格
plt.plot(xnew, ynew1, linewidth='1', color='cyan',
         linestyle='--', marker='o', label='Goodware')
plt.plot(xnew, ynew2, 'b', linestyle='-.', marker='.', label='Critroni')
plt.legend()  # 显示上面的label
plt.xlabel('feature dimension')
plt.ylabel('Accurancy')
plt.axis([500, 2500, 0, 1])  # 设置坐标范围axis([xmin,xmax,ymin,ymax])

plt.show()



#画图
# plt.title('Compare cosx with sinx')  #标题
#plt.plot(x,y)
#常见线的属性有：color,label,linewidth,linestyle,marker等
# plt.grid() # 生成网格
# plt.plot(x, y0, linewidth='1',color='cyan', linestyle='--', marker='o', label='Goodware')
# # 'b'指：color='blue'
# plt.plot(x, y1, 'b', linestyle='-.', marker='.',label='Critroni')
# plt.legend()  #显示上面的label
# plt.xlabel('feature dimension')
# plt.ylabel('Accurancy')
# plt.axis([500, 2500, 0, 1])#设置坐标范围axis([xmin,xmax,ymin,ymax])
# # plt.xlim(500,2000)#仅设置x轴坐标范围
# # plt.ylim(0,1)#仅设置y轴坐标范围
# plt.show()


# # 3. 绘制平滑曲线
# from scipy.interpolate import spline
# # 插值法，50表示插值个数，个数>=实际数据个数，一般来说差值个数越多，曲线越平滑
# x_new = np.linspace(min(x), max(x), 50)

# y_smooth = spline(x, y1, x_new)

# plt.plot(x_new, y_smooth)

# plt.show()

###
# https://www.cnblogs.com/caiyishuai/p/9720195.html
# https://blog.csdn.net/u013317445/article/details/83826717
```
