```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *  # 支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']
# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 250像素/每英寸
plt.figure(figsize=(10, 8), dpi=230)

# 再创建一个规格为 1 x 1 的子图
plt.subplot(1, 1, 1)

# 柱子总数
N = 11
# 包含每个柱子对应值的序列
values = (1941, 391, 22, 328, 1387, 2928, 1063, 394, 198,117,343)

# 包含每个柱子下标的序列
index = np.arange(N)

# 柱子的宽度
width = 0.3

# 绘制柱状图, 每根柱子的颜色为紫罗兰色
p2 = plt.bar(index, values, width, color="#87CEFA")

# 设置横轴标签
plt.xlabel('样本字符出现次数',fontsize=12)
# 设置纵轴标签
plt.ylabel('出现对应次数的汉字数量', fontsize=12)

# 添加标题
# plt.title('文本检测对比图')

# 添加纵横轴的刻度
plt.xticks(index, ('<10', '<20', '<30',
                   '<40', ' <50', '<100', '<200', '<300', ' <400', '<500', '>500'))
plt.yticks(np.arange(0, 4001, 1000))

# 添加图例
plt.legend(loc="upper right")
for a, b in zip(index, values):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.show()
```

```python
# 显示数据集分类
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  


# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 250像素/每英寸
plt.figure(figsize=(10, 6), dpi=230)

x = np.array(['Truck','Auxiliary','Car','Excavator','Widebody','Pedestrian'])
y = np.array([2432, 364, 362, 357, 299, 130])

plt.bar(x, y,  color = ["#4CAF50","red","hotpink","#556B2F","green","blue"])

#使用text显示数值  
for a,b in zip(x,y):  
 plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=11)

plt.xlabel("Label")  #设置X轴Y轴名称  
plt.ylabel("Number")  
# plt.title("bar chart")
plt.style.use('seaborn-darkgrid') # 设置画图的风格 
# 保存图片
fig=plt.gcf()
fig.savefig('./mine.jpg')

plt.show()
```
