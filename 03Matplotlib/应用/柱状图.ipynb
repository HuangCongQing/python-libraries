```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def auto_label(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height), # put the detail data
                    xy=(rect.get_x() + rect.get_width() / 2, height), # get the center location.
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


def auto_text(rects):
    for rect in rects:
        ax.text(rect.get_x(), rect.get_height(), rect.get_height(), ha='center', va='bottom')


labels = ['RS-LiDAR-16', 'RS-Bpearl']
men_means = [2859, 2935]
women_means = [2679, 2783]

index = np.arange(len(labels))
width = 0.4

fig, ax = plt.subplots()
rect1 = ax.bar(index - width / 2, men_means, color ='lightcoral', width=width, label ='Slope')
rect2 = ax.bar(index + width / 2, women_means, color ='springgreen', width=width, label ='Ground')
# rect3 = ax.bar(index + width / 2, means, color ='springgreen', width=width, label ='means')

# ax.set_title('Scores by gender')
ax.set_xticks(ticks=index)
ax.set_xticklabels(labels)
ax.set_ylabel('Number')
ax.set_xlabel('LiDAR Type')

ax.set_ylim(0, 4000)
# auto_label(rect1)
# auto_label(rect2)
auto_text(rect1)
auto_text(rect2)
# auto_text(rect3)

ax.legend(loc='upper right', frameon=False)
fig.tight_layout()
plt.savefig('2.jpg', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *  # 支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']
# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 250像素/每英寸
plt.figure(figsize=(10, 8), dpi=230)

# 再创建一个规格为 1 x 1 的子图
plt.subplot(1, 1, 1)

# 柱子总数
N = 11
# 包含每个柱子对应值的序列
values = (1941, 391, 22, 328, 1387, 2928, 1063, 394, 198,117,343)

# 包含每个柱子下标的序列
index = np.arange(N)

# 柱子的宽度
width = 0.3

# 绘制柱状图, 每根柱子的颜色为紫罗兰色
p2 = plt.bar(index, values, width, color="#87CEFA")

# 设置横轴标签
plt.xlabel('样本字符出现次数',fontsize=12)
# 设置纵轴标签
plt.ylabel('出现对应次数的汉字数量', fontsize=12)

# 添加标题
# plt.title('文本检测对比图')

# 添加纵横轴的刻度
plt.xticks(index, ('<10', '<20', '<30',
                   '<40', ' <50', '<100', '<200', '<300', ' <400', '<500', '>500'))
plt.yticks(np.arange(0, 4001, 1000))

# 添加图例
plt.legend(loc="upper right")
for a, b in zip(index, values):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.show()
```
