```python
    
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *  # 支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(10, 8), dpi=200)

names = ['0', '<=1', '<=1/2', '<=1/3', '<=1/4',
         '<=1/5', '<=1/6', '<=1/7', '<=1/8', '<=1/9',
         '<=1/10', '<=1/11', '<=1/12', '<=1/13', '<=1/14',
         '<=1/15','<=1/16', '<=1/17', '<=1/18', '<=1/19', '<=1/20']
x = range(len(names))
y = [0, 0.02, 0.06, 0.3, 0.06, 0.1, 0.090, 0.080, 0.07, 0.05, 0.02,
     0.03, 0.03, 0.02, 0.03, 0.05, 0.04, 0.02, 0.01, 0.01, 0.01]
#plt.plot(x, y, 'ro-')
#pl.xlim(-1, 11)  # 限定横轴的范围
# pl.ylim(0, 0.35)  # 限定纵轴的范围
plt.plot(x, y, marker='o', mec='r', mfc='w', label=u'对应宽高比的比例')

plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel("样本文本行宽高比", fontsize=12)  # X轴标签
plt.ylabel("对应宽高比例", fontsize=12)  # Y轴标签
# plt.title("A simple plot")  # 标题

plt.show()
```
