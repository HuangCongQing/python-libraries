## 绘制多条折线图

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]  # x轴数据
y1 = [1, 4, 9, 16, 25]  # 第一条折线图的y轴数据
y2 = [1, 2, 3, 4, 5]  # 第二条折线图的y轴数据

plt.plot(x, y1, label='Line 1')  # 绘制第一条折线图
plt.plot(x, y2, label='Line 2')  # 绘制第二条折线图

plt.legend()  # 添加图例
plt.xlabel('X-axis')  # 添加x轴标签
plt.ylabel('Y-axis')  # 添加y轴标签

plt.show()
```
