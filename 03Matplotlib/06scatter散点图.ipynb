## 单张散点图

```python
# matplotlib和pyplot的惯用别名分别是mpl和plt

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# 通过rcParams设置全局横纵轴字体大小
mpl.rcParams['xtick.labelsize'] = 24
mpl.rcParams['ytick.labelsize'] = 24

np.random.seed(42)
# print(np.random.seed(42))

# x轴的采样点
x = np.linspace(0, 5, 100)# 等差数列
# print('x数据：\n',x)
# 通过下面曲线加上噪声生成数据，所以拟合模型就用y了……
y = 2*np.sin(x) + 0.3*x**2
y_data = y + np.random.normal(scale=0.3, size=100)


def show_scatter(x, y, label=None):
    # figure()指定图表名称
    plt.figure('data', figsize=(8, 8))
    # scatter可以更容易地生成散点图
    label = y
    plt.scatter(x, y, marker='o', c=label, cmap='coolwarm')

    # 将当前figure的图保存到文件result.png
    plt.savefig('result.png')

    # 一定要加上这句才能让画好的图显示在屏幕上
    plt.show()

show_scatter(x,y)







```

### 根据label变化(and显示颜色和标签对应关系)
通过 `plt.scatter` 绘制散点图，并且希望根据标签的数值大小来调整点的颜色深浅。您可以使用 `c` 参数来指定颜色，以及 `cmap` 参数来选择颜色映射。下面是一个示例代码：


在这个示例中，我们使用了随机生成的数据，其中 `labels` 数组包含了标签的数值。通过将 `c` 参数设置为 `labels`，并选择适当的颜色映射（在这里我们使用了 'viridis'），就可以根据标签数值来调整点的颜色深浅。点的大小由 `s` 参数控制。最后，通过 `plt.colorbar()` 添加一个颜色标注，以显示颜色和标签值之间的对应关系。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一些示例数据
x = np.random.rand(50)
y = np.random.rand(50)
labels = np.random.randint(1, 11, size=50)  # 随机生成标签数值（范围在1到10之间）

# 绘制散点图，并根据标签数值调整颜色深浅
# plt.scatter(x, y, c=labels, cmap='viridis', s=50)  # 使用'viridis'颜色映射，s参数控制点的大小
plt.scatter(x, y, c=labels, cmap='rainbow', s=50) 

# 添加颜色标注
cbar = plt.colorbar()
cbar.set_label('Label Values')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Label-Colored Points')
plt.show()
```

## 多张散点图
* https://blog.csdn.net/qq_37851620/article/details/100642566

```python
# 生成样本点
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=500,
                           n_features=10,
                           n_classes=5,
                           n_informative=4,                          
                           random_state=0)
```

### 1. 指定参数c

```python
# # 默认绘图
# import matplotlib.pyplot as plt
# plt.figure(figsize=(8, 8))
# plt.scatter(X[:, 0], X[:, 1], marker='o')

# 1. 指定参数c
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], marker='o', c='g')
```

###  2. c=y，设置cmap

```python
# 2. c=y，设置cmap
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='coolwarm')
```

### 多张图

```python
plt.figure(figsize=(12, 8))
plt.subplot(231)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='plasma')
plt.subplot(232)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='Oranges')
plt.subplot(233)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='summer')
plt.subplot(234)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='Spectral')
plt.subplot(235)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='Set1')
plt.subplot(236)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='rainbow')
plt.show()
```
