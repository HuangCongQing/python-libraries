```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot data
# Series
data = pd.Series(np.random.randn(1000),index=np.arange(1000))# index默认
data = data.cumsum()# 累加
print(data)
data.plot()
```

```python
# DataFrame
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000), columns=list("ABCD"))
print
data = data.cumsum()
print(data.head())# 默认head()显示前五个
data.plot()
plt.show()
# plt.scatter(x=,y=)
```
